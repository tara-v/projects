```python
# standardizing column names

def standardize_cols(name):
    name = name.replace('__', '_')
    name = name.replace(' ', '_')
    name = name.replace('/', '_')
    name = name.replace('(', '')
    name = name.replace(')', '')
    name = name.replace('%', 'pct')
    name = name.replace('<', 'below')
    name = name.replace('>', 'above')
    name = name.replace('$', 'usd')
    name = name.replace('?', '')
    ame = name.replace("'", '')
    name = name.lower()
    return name


standardized = []

for column in main.columns:
    standardized.append(standardize_cols(column))

main.columns = standardized
```

```python
# filtering to only CA data

california = main[main.state_territory == 'California']
california.head()
```

```python
#exporting cleaned primary dataset
california.to_csv('ca_main.csv', index=False)
```

### Supplemental Dataset 1 - IPUMS Land Coverage

- ***main dataset key***: 'census_tract_2010_id'
- ***supplemental dataset #1***: IPUMS land coverage 
	- key: GISJOIN
	- Documentation: The GISJOIN field contains NHGIS-standard tract identifiers:
		- A county GISJOIN is a concatenation of:
		- "G"
		- State NHGIS code: 3 digits (FIPS + "0")
		- County NHGIS code: 4 digits (FIPS + "0")
		- Census tract code: 6 digits
	
- ***Steps to join***:
	- clean GISJOIN 
	- remove first 3 characters of Census Tract ID from main dataset


*Notes*: 
- FIPS codes reference: https://transition.fcc.gov/oet/info/maps/census/fips/fips.txt 
- Census geoid reference: https://www.census.gov/programs-surveys/geography/guidance/geo-identifiers.html 
	- state: 2 digits (06) 
	- county: 3 digits (06 + 109) -> 5 digits total 
	- census tract: 6 digits (06 + 109 + 390087) -> 11 digits total (main dataset, IPUMS 2 datasets)

```python
# import

land = pd.read_excel('../data/tract_IPUMS_landcoverage_2015.xlsx')
land.head()
```

```python
# cleaning GISJOIN column to match census tract GEOID

land['GISJOIN'] = land['GISJOIN'].astype(str)
land['merge_id'] = land['GISJOIN'].str[1:3] + \
    land['GISJOIN'].str[4:7] + land['GISJOIN'].str[8::]
land.head()
```

```python
# removing land coverage variables with high missingness

land_maincols = land.loc[:, ['AREA21', 'AREA22', 'AREA23',
                             'AREA24', 'PROP21', 'PROP22', 'PROP23', 'PROP24', 'merge_id']]
land_maincols.shape
```

```python
# adding missing 0 to beginning of CA census tract FIPS code

california['merge_id'] = '0' + california['census_tract_2010_id'].astype(str)
california.head()
```

```python
land.shape
```

```python
land_ca = land[land['merge_id'].str[:2] == '06']
land_ca.shape
```

```python
# left join with primary dataset

ca_land = pd.merge(california, land_maincols, on='merge_id', how='left')
ca_land.shape
```

```python
pd.set_option('display.max_columns', None)
ca_land.head()
```

### Supplemental Dataset 2 - IPUMS Temperature

```python
# import
temp = pd.read_excel('../data/county_IPUMS_temp_2015.xlsx',
                     sheet_name='US_county_2015_annualtmean')
temp.head()
```

```python
# cleaning GISJOIN column to match census tract GEOID

temp['GISJOIN'] = temp['GISJOIN'].astype(str)
temp['merge_id_county'] = temp['GISJOIN'].str[1:3] + temp['GISJOIN'].str[4:7]
temp.head()
```

```python
# converting land census tract ID to string

ca_land['census_tract_2010_id'] = ca_land['census_tract_2010_id'].astype(str)
```

```python
# adding county merge ID to ca_land

ca_land['merge_id_county'] = ca_land['merge_id'].str[:5]
ca_land['census_tract_2010_id'] = '0' + ca_land['census_tract_2010_id']
ca_land.head()
```

```python
ca_land.shape
```

```python
# data is included for multiple years - just join latest data from 2014 only

temp_2014 = temp[temp['TIME'] == 2014]
temp_2014.info()
```

```python
# check that temp data exists for all counties in california

temp_2014_ca = temp_2014[temp_2014['merge_id_county'].str[:2] == '06']
temp_2014_ca.shape
```

```python
# filter to relevant columns

temp_2014_maincols = temp_2014.loc[:, [
    'MIN', 'MAX', 'MEAN', 'STD', 'TIME', 'merge_id_county']]
temp_2014_maincols.shape
```

```python
# merge using left join
ca_land_temp = pd.merge(ca_land, temp_2014_maincols,
                        on='merge_id_county', how='left')
ca_land_temp.shape
```

```python
# converting temps into fahrenheit

ca_land_temp.loc[:, 'MEAN_F'] = (ca_land_temp['MEAN']*(9/5))+32
ca_land_temp.head()
```

```python
# standardizing column names

standardized = []

for column in ca_land_temp.columns:
    standardized.append(standardize_cols(column))

ca_land_temp.columns = standardized
```
